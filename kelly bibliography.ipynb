## Kelly Criterion - a limited bibliography

The following are many of the most important papers to date on the Kelly Criterion, along with a few where authors tried to extend the method.  In most cases, I tried to summarize the main finding or findings.  In a couple, I highlighted where the authors might have gone astray, hopefully in a tactful way.  While I view all research as good, understanding where more work can be done is useful.  Often, find that work lies in looking for potential weaknesses in the existing literature.

<u>Kelly, J. L. (1956). A new interpretation of information rate. the bell system technical journal, 35(4), 917-926. <a href="https://www.princeton.edu/~wbialek/rome/refs/kelly_56.pdf">(link)</a></u>

<i><font color = 'green'>Kelly's paper was the original one that kicked off an entire literature of bet sizing.  For such a simple idea, it is remarkably hard to read, mainly because he couched his results in an information theory and bits being sent down a wire.  It's worth reading once to get the main ideas but clearer papers came later.</i></font>

<u>Breiman, L. (1961). Optimal gambling systems for favorable games. The Kelly Capital Growth Investment Criterion, 47-60. <a href="http://www-stat.wharton.upenn.edu/~steele/Resources/FTSResources/KellyBreiman/Breiman61.pdf">(link)</a></u>

<i><font color = 'green'>I frankly looked to this paper to see if there was more justification to using the logarithmic wrapper, turning Kelly into a utility statement instead of a wealth maximization statement.  In Breiman's case, using the log formulation allows him to use martingale theory to prove some interesting but somewhat abstract theorems.  Later papers credit Breiman for proving that optimzing the expectation of the log of wealth is desirable.  Others credit Latane (1959) for proving that optimizing the geometric mean of wealth is superior.</i></font>

<u>Thorp, E. O. (1969). Optimal gambling systems for favorable games. Revue de l'Institut International de Statistique, 273-293.</u>

<i><font color = 'green'>Thorp's discussion is wide ranging at first but all of his calculations revolve around stocks and options.  He makes a case that Kelly outperforms Markowitz optimization but the motivating example is somewhat unrealistic, leaving the conclusion uncertain.  His final example explores investing in two independent bets (and a risk free asset) and shows that there is no optimal portfolio.</i></font>

<u>Thorp, E. O. (1975). Portfolio choice and the Kelly criterion. In Stochastic optimization models in finance (pp. 599-619). Academic Press.</u>

<i><font color = 'green'>This paper is a little harder to interpret.  Much of it is a counter-argument to Paul Samuelson who stated that lognormal utility is not a good utility function for modeling.  Thorp disagrees and shows that the second order Taylor expansion is very close to Markowitz optimation.  This is a strange argument because that is a basic understanding of financial economics.  He also says that using Kelly for portfolio optimization led him to achieve a 25% annual return.  The work was supposed to be in his 1967 book, Beat the Market.</i></font>

<u>Browne, S., & Whitt, W. (1996). Portfolio choice and the Bayesian Kelly criterion. Advances in Applied Probability, 28(4), 1145-1176. <a href="https://www.columbia.edu/~ww2040/PortfolioChoice96.pdf">(link)</a></u>

<i><font color = 'green'>This is an interesting paper.  I think it is the first time a random walk process was applied to the Kelly Criteria to derive a continuous time betting process.  The result is a return process that achieves a geometric brownian motoion.  The more interesting concept is to assume uncertainty around the probability of success estimate and to use Bayesian analysis to infer a more robust Kelly fraction.</i></font>

<u>Ziemba, B. (2006). Good and bad properties of the Kelly criterion. The Best of Wilmott, 65. </u>

<i> <font color = 'green'>This paper was republished almost exactly as-is with extra authors in 2011. </i> </font>


<u>Thorp, E. O. (2008). The Kelly criterion in blackjack sports betting, and the stock market. In Handbook of asset and liability management (pp. 385-428). North-Holland. <a href="http://www.eecs.harvard.edu/cs286r/courses/fall12/papers/Thorpe_KellyCriterion2007.pdf">(link)</a></u>

<i> <font color = 'green'>Much of this article is a rehash of mathematical results from previous papers.  One interesting anecdote is that the author and a Ph.D. candidate bet on sports, turning \$50,000 into \$123,000 over four and a half months in 1994.  Other notable derivations are (1) two independent bets, showing that the optimal bet sizes are not independent, (2) a intro to correlated bets though the analysis is brief and not fleshed out, and (3) some application to stock portfolios, albeit with only two stocks. There is a nice multivariate derivation but the symbols are unclear and require significant analysis to understand.  In the end, as much as I liked the paper, it is hard to read and seems like a bit of advertising for Thorp's investment vehicle.</i></font>

<u>MacLean, L. C., Thorp, E. O., & Ziemba, W. T. (2011). Introduction to the Good and Bad Properties of Kelly. In THE KELLY CAPITAL GROWTH INVESTMENT CRITERION: THEORY and PRACTICE (pp. 459-464).</u>

<i> <font color = 'green'>This is a very short but interesting discussion of the pros and cons of the Kelly Criterion.</u></font>

<u>Ziemba, W. T. (2012). Response to Paul A Samuelson letters and papers on the Kelly capital growth investment strategy. Available at SSRN 2161068. <a href="https://papers.ssrn.com/sol3/papers.cfm?abstract_id=2161068">(link)</a></u>

<i> <font color = 'green'>This is a formal paper based on discussions that Ziemba had with Samuelson in 2008.  The main point he makes is that Kelly almost surely leads to higher long term wealth than any other betting strategy.  He admits that it is based on logarithmic utility (or CRRA utility) and that other solutions are possible with other utility functions.</i></font>

<u>Baker, R. D., & McHale, I. G. (2013). Optimal betting under parameter uncertainty: Improving the Kelly criterion. Decision Analysis, 10(3), 189-199. <a href = "https://www.researchgate.net/profile/Rose-Baker-6/publication/262425087_Optimal_Betting_Under_Parameter_Uncertainty_Improving_the_Kelly_Criterion/links/54abd3ea0cf2ce2df669133d/Optimal-Betting-Under-Parameter-Uncertainty-Improving-the-Kelly-Criterion.pdf">(link)</a></u>

<i> <font color = 'green'>Baker et al show that, in the presence of uncertainty around the probabilty of success, shrinking the Kelly fraction leads to higher median returns and higher utility, even if the full Kelly fraction still outperforms for average wealth growth. </i> </font>


<u>Ziemba, W. T. (2015). A response to Professor Paul A. Samuelson's objections to Kelly capital growth investing. Journal of Portfolio Management, 42(1), 153.</u>

<i> <font color = 'green'>Ziemba's 2012 paper updated and published in the JPM. </i></font>

<u>Khanna, N. (2016). The Kelly criterion and the stock market. <a href="http://www.edwardothorp.com/wp-content/uploads/2016/11/TheKellyCriterionAndTheStockMarket.pdf">(link)</a></u>

<i> <font color = 'green'>Khanna demonstrates how a random walk for returns can be used to simulate a continuous time progression.  Instead of a full derivation, he stops after setting up the problem, taking the second order Taylor series, and shows that the Kelly fraction is equivalent to the Markowitz investment amount in a universe without correlations.  It's a very short paper. </i> </font>

<u>Peterson, Z. (2017). Kelly's criterion in portfolio optimization: a decoupled problem. arXiv preprint arXiv:1710.00431. <a href="https://arxiv.org/pdf/1710.00431">(link)</a></u>

<i> <font color = 'green'>Peterson adapts Kelly to portfolio analysis and uses numerical analysis to generate solutions.  I think the derivation is incorrect and will be putting up a different solution in about two weeks.  His main takeaway seems to be that the results don't significantly deviate from Markowitz though that might simply be a statement that his results are reasonable, not that they aren't different.</i> </font>

<u>Chu, D., Wu, Y., & Swartz, T. B. (2018). Modified kelly criteria. Journal of Quantitative Analysis in Sports, 14(1), 1-11. <a href="http://www.sfu.ca/~tswartz/papers/kelly.pdf">(link)</a></u>

<i> <font color = 'green'>Chu et al use Bayes analysis to estimate Kelly fractions when the probablity estimate is uncertain.  This differs from Baker et al (2013) by expanding on the theoretical analysis a bit more.</i> </font>

<u>Lo, A. W., Orr, H. A., & Zhang, R. (2018). The growth of relative wealth and the Kelly criterion. Journal of Bioeconomics, 20, 49-67. <a href="https://dspace.mit.edu/bitstream/handle/1721.1/114640/10818_2017_9253_ReferencePDF.pdf?sequence=2&isAllowed=y">(link)</a></u>

<i> <font color = 'green'>It might seem that this paper is about the Kelly Criterion but it is actually about investor behavior when investors are envious of others.  In this framework, absolute wealth no longer matters, only wealth relative to others.  The authors show that the Kelly fraction is correct for absolute growth but optimality deviates in some situations when the investor is actually trying to maximize relative wealth. </i> </font>

<u> Matej, U., Gustav, Š., Ondřej, H., & Filip, Ž. (2021). Optimal sports betting strategies in practice: an experimental review. IMA Journal of Management Mathematics, 32(4), 465-489. <a href="https://www.researchgate.net/profile/Gustav-Sir/publication/353344291_Optimal_sports_betting_strategies_in_practice_an_experimental_review/links/60faccbd2bf3553b29086bde/Optimal-sports-betting-strategies-in-practice-an-experimental-review.pdf">(link)</a></u>

<i><font color = 'green'>An experimental paper comparing several betting strategies.  The authors make the case that fractional Kelly (where the betting fraction is less than the Kelly fraction) is the best strategy.</i></font>

<u>Tran, S., & Verhoeven, P. (2021). Kelly criterion for optimal credit allocation. Journal of Risk and Financial Management, 14(9), 434.</u>

<i> <font color = 'green'>A topic near and dear to my heart - optimal credit lending strategies.  The paper uses the Kelly fraction to determine how much to lend.  Ultimately, the paper seems more like a test of probability of default (PD) estimators than it does of Kelly, it is still good to see work in this area.</i></font>
